# Data Memo: Formula 1 Race Outcome Prediction

### Overview of Dataset

For my final project, I plan to use both historical Forumla 1 data and read-time race data to build a machine learning model to be able to predict race outomes. I will use the historical data to train/test the model, and use real time will be useful when I turn this project into an interactive website.

I plan to use the [Formula 1 World Championship (1950 - 2024)](https://www.kaggle.com/datasets/rohanrao/formula-1-world-championship-1950-2020) data set from Kaggle. It includes race results, drivers, lap times, and other relavent race-related data that would be helpfull to train the model to create accurate predictions. This data set includes many csv files, which can be joined with keys like raceID and driverID.

For real time data I plan to use [FastF1](https://github.com/theOehrly/Fast-F1). This package allows me to pull real time race schedules, current results, and other live racing data. I will use this package to test if the model can create accurate predictions and for future implementation. 

An observation in my training dataset will represent one driver in one race. Predictors include grid position, qualifying position, driver/team, circuit, weather conditions, driver standings, and other performace metrics. The outcome variable would be driver finishing position and/or whether the driver finishes on the podium (top 3).

The [Formula 1 World Championship (1950 - 2024)](https://www.kaggle.com/datasets/rohanrao/formula-1-world-championship-1950-2020) dataset includes both categorical and numerical variables.

    Numerical: grid position, finishing position, points, laps completed, qualifying rank, and standings points

    Categorical: driver name, constructor/team, circuit, race name, nationality, and race status


Older races may be prone to missing data, and to handle this, I may train the model on more recent races (2012 onwards). I may also fill in missing data by hand or create indicators.

### Research Question

For this project I will be predicting Formula 1 race outcomes based on information about driver performance and pre-race data. The main question I plan to pursue is **whether a driver's chance of finishing on the podium can be predicted accurately given parameters like qualifying position, starting grid position, constructor/team, driver, circuit, and recent performance**.

My response variable would be podium finish. I will create a binary variable where 1 means the driver finished in the podium, and 0 if they did not. I am also interested in making the predicted probability of each driver and their position.
